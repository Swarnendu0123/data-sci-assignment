Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.

````python
A = [1,2,3,4,5,6,7,8,9]
B = [2,4,6,8,10,12,14,16,18]
C = [1, 4, 9, 16, 25, 36, 49, 64, 81]
````

````python
def re_index(df):
  new_index = pd.Index(range(1, len(df)*2 + 1, 2))
  df = df.set_index(new_index)
  return df
````

````python
df1 = pd.DataFrame({"A": A, "B":B, "C":C})
````

````python
df1
````

````python
df1 = re_index(df1)
````

Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.
The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.

````python
df2 = pd.DataFrame({"e_mail" : [
  "johndoe@gmail.com",
  "sarahsmith@gmail.com",
  "robertjones@gmail.com",
  "emilywilson@gmail.com",
  "chrisbrown@gmail.com",
  "lindajackson@gmail.com",
  "markthompson@gmail.com",
  "lisasmith@gmail.com",
  "kevinmiller@gmail.com",
  "amandajohnson@gmail.com"
]}
)
````

````python
def extract_username(df2):
  usernames = []
  for i in range(len(df2["e_mail"])):
    usernames.append(df2["e_mail"][i].split("@")[0])
  return usernames
````

````python
username = extract_username(df2)
df2["user_name"] = username
df2
````

Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.
For example, if df contains the following values:
```
Date
0 2023-01-01
1 2023-01-02
2 2023-01-03
3 2023-01-04
4 2023-01-05
```
Your function should create the following DataFrame:
```
        Date Weekday
0 2023-01-01 Sunday
1 2023-01-02 Monday
2 2023-01-03 Tuesday
3 2023-01-04 Wednesday
4 2023-01-05 Thursday
```
The function should return the modified DataFrame.

````python
dates = ["2023-01-01","2023-01-02","2023-01-03","2023-01-04","2023-01-05"]
df5 = pd.DataFrame({"Date":dates})
````

````python
df5
````

````python
from datetime import datetime
def add_week_day(df5):
  Weekday = []
  for i in range(len(df5)):
    year = int(df5["Date"][i].split("-")[0])
    month = int(df5["Date"][i].split("-")[1])
    day = int(df5["Date"][i].split("-")[2])
    # Createing a datetime object
    x = datetime(year, month, day)
    days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    # Geting the name of the weekday
    weekday_name = days[x.weekday()]
    Weekday.append(weekday_name)

  df5["Weekday"]= Weekday
````

````python
add_week_day(df5)
df5
````

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

````python
# Sample DataFrame with 'Date' column containing timestamps
data = {'Date': ['2023-01-15', '2023-02-10', '2023-01-05', '2023-02-20']}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])  # Convert the 'Date' column to datetime format
````

````python
df
````

````python
def select_rows_in_date_range(dataframe, start_date, end_date):
    mask = (dataframe['Date'] >= start_date) & (dataframe['Date'] <= end_date)
    selected_rows = dataframe[mask]
    return selected_rows
````

````python
# Define the start and end date for your range
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-01-31')
````

````python
# Use the function to select rows within the specified date range
selected_rows = select_rows_in_date_range(df, start_date, end_date)

# Print the selected rows
print(selected_rows)
````
