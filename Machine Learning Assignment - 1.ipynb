## Unsupervised ML: we perform some oprations on unlebeled dataset and try to catagorise the data.
## Semi-Supervised ML: a model is trained on a dataset that contains both labeled and unlabeled data. In traditional supervised learning, the algorithm learns from labeled data, where input features are paired with corresponding output labels. Unsupervised learning, on the other hand, involves learning patterns and structures from unlabeled data.

# Ans 6
## Train dataset: The training data is the portion of the dataset that is used to train the machine learning model.
## Validation Dataset: The validation data is a subset of the dataset that is used to tune hyperparameters and fine-tune the model during the training process. It is not used for training the model, but rather for selecting the best model from different variations
## Test Dataset: The test data is a separate portion of the dataset that is not used during the training process. It is used to evaluate the performance of the trained model
